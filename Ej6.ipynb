```python
import random
import matplotlib.pyplot as plt
from itertools import accumulate
import numpy as np
lam = 3
tmax = 24*3600*7
registro = []
for i in range(tmax):
  registro.append(1 if random.random() < (lam / 3600) else 0)
f = plt.figure(dpi=300)
ax = f.add_subplot(111)
ax.set_aspect(600)
c = list(accumulate(registro))
plt.plot(c)
plt.title("Arribos al aeropuerto")
plt.xlabel("Tiempo (en horas)")
plt.ylabel("Cantidad de arribos")
plt.xticks(np.arange(0, tmax+1, 3600*24), [i*24 for i in range(8)], fontsize=6)
plt.yticks(fontsize=6)
llegadas = [i for i in range(len(registro)) if registro[i] == 1]
tiempos = [llegadas[i+1] - llegadas[i] for i in range(len(llegadas) - 1)]
plt.figure(dpi=300)
plt.hist(tiempos, bins=np.arange(0, 7200, 600))
aa = np.arange(0, 7201, 600)
plt.xticks(aa, [i*10 for i in range(len(aa))])
plt.title("Tiempo entre arribos")
plt.xlabel("Tiempo (en minutos)")
plt.ylabel("Frecuencia")
np.mean(tiempos)/3600
```
